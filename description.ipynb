# Гистограмма

Гистограмма — способ представления табличных данных в графическом виде — в виде столбчатой диаграммы.

Суть данного метода заключается в том, что мы разбиваем область определения значений нашей выборки на r интервалов, после чего считаем для каждого интервала количество попаданий в него. Далее графически над каждым интервалом надстраиваем столбец, высота которого соответствует количеству принадлежащих ему значений выборки.

Гистограмма используется в качестве наглядного изображения выборочной плотности. В нашей задаче такой постановки задачи не было, поэтому мы использовали её, чтобы увидеть распределение значений (например, количества добавленных или обновлённых записей) на временной прямой.

# Диаграмма

Диаграмма — графическое представление данных линейными отрезками или геометрическими фигурами, позволяющее быстро оценить соотношение нескольких величин.

В нашей задачи анализа мы распространённо использовали данный метод, ведь сырые выводы, сделанные программой, не самым лучшим образом иллюстрируют результат, как для нас самих, так и для конечных обозревателей. Поэтому в нашей задачи мы использовали гистограммы, диаграммы, графики, а также интерактивные карты, на которых размер кружка-города зависел от количества отелей в нём.

# Линейная регрессия

Линейная регрессия — модель зависимости переменной x от одной или нескольких других переменных (факторов, регрессоров, независимых переменных) с линейной функцией зависимости

Суть метода сводится к тому, что мы получаем и используем уравнение вида:

y = a + b1x1 + b2x2... +bn*xn

И используем его для того, чтобы предсказывать значение y на основе вектора значений факторов. Это одна из простейший моделей предсказания.

Вычислять данные коэффициенты по методу наименьших квадратов: мы хотим, чтобы отклонение предсказанных значений от реальных было минимальным. Составляем функцию - сумму квадратов разностей значений, берём частные производные, решаем систему уравнений. В итоге получаем коэффициенты, зависящие от коэффициентов корреляции - понятное дело, ведь корелляция отображает линейную зависимость между величинами.

# Логистическая регрессия

Логистическая регрессия — статистическая модель, используемая для прогнозирования вероятности возникновения некоторого события путём его сравнения с логистической кривой. Эта регрессия выдаёт ответ в виде вероятности бинарного события (1 или 0).

Логистическая регрессия применяется для прогнозирования вероятности возникновения некоторого события по значениям множества признаков. Для этого вводится так называемая зависимая переменная y, принимающая лишь одно из двух значений — как правило, это числа 0 (событие не произошло) и 1 (событие произошло), и множество независимых переменных (также называемых признаками, предикторами или регрессорами) — вещественных x{1},x{2},...,x{n}, на основе значений которых требуется вычислить вероятность принятия того или иного значения зависимой переменной. Как и в случае линейной регрессии, для простоты записи вводится фиктивный признак x{0}=1

Эта модель часто применяется для решения задач классификации — объект x можно отнести к классу y=1, если предсказанная моделью вероятность P{y = 1 | x} > 0,5 и к классу y=0 в противном случае. Получающиеся при этом правила классификации являются линейными классификаторами.


# Random forest

Random forest — алгоритм машинного обучения, предложенный Лео Брейманом и Адель Катлер, заключающийся в использовании комитета решающих деревьев. 

Это универсальный алгоритм машинного обучения, суть которого состоит в использовании ансамбля решающих деревьев. Само по себе решающее дерево предоставляет крайне невысокое качество классификации, но из-за большого их количества результат значительно улучшается. Также это один из немногих алгоритмов, который можно использовать в абсолютном большинстве задач.

Тренировка деревьев происходит независимо друг от друга (на разных подмножествах), что не просто решает проблему построения одинаковых деревьев на одном и том же наборе данных, но и делает этот алгоритм весьма удобным для применения в системах распределённых вычислений.

# Кластеризация

Кластеризация — задача группировки множества объектов на подмножества (кластеры) таким образом, чтобы объекты из одного кластера были более похожи друг на друга, чем на объекты из других кластеров по какому-либо критерию.

Задача кластеризации относится к классу задач обучения без учителя.

Критерии кластеризации определяет человек, а не алгоритм, — этим она отличается от классификации. Этот метод машинного обучения (Machine Learning) часто применяют в различных неструктурированных данных — например если нужно автоматически разбить коллекцию изображений на мини-группы по цветам.
